```python
import pandas as pd
import numpy as np
import random as rnd
from sklearn.preprocessing import LabelEncoder
# visualization
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

# machine learning
from sklearn.metrics import mean_squared_error, r2_score
from sklearn import linear_model
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC, LinearSVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.linear_model import Perceptron
from sklearn.linear_model import SGDClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestRegressor
```

```python
#Read the csv into dataframes
train_dataframe = pd.read_csv('train_2016_v2.csv' , parse_dates=["transactiondate"])

properties_dataframe = pd.read_csv('properties_2016.csv')  

test_dataframe = pd.read_csv('sample_submission.csv')
test_dataframe= test_dataframe.rename(columns={'ParcelId': 'parcelid'}) #to make it easy for merging
```

```python
print("Training Size:" + str(train_dataframe.shape))

print("Property Size:" + str(properties_dataframe.shape))

print("Sample Size:" + str(test_dataframe.shape))
```

```python
print("Columns in Training:")
print(train_dataframe.columns.values)
print("Columns in Properties:")
print(properties_dataframe.columns.values)
```

```python
#for faster processing convert datatypes to required format
for col, dtype in zip(properties_dataframe.columns, properties_dataframe.dtypes):

    if dtype == np.float64:       

        properties_dataframe[col] = properties_dataframe[col].astype(np.float32)

    if dtype == np.int64:

        properties_dataframe[col] = properties_dataframe[col].astype(np.int32)

for column in test_dataframe.columns:

    if test_dataframe[column].dtype == int:

        test_dataframe[column] = test_dataframe[column].astype(np.int32)

    if test_dataframe[column].dtype == float:

        test_dataframe[column] = test_dataframe[column].astype(np.float32)
```

Question 3: Linear Regression
Flow:
1. Creating test and train datasets from properties data
2. Pre processing data
3. Split data and drop unwanted columns
4. Create, fit and score model
5. Predict logerror for entire dataset and write to csv
6. Sample test set and predict values
7. Calculate mean error and r2 values

```python
#Linear Regression
#Merging test and train datasets
df_train = train_dataframe.merge(properties_dataframe, how='left', on='parcelid')
df_test = test_dataframe.merge(properties_dataframe, how='left', on='parcelid')
```

```python
#Clean data by filling 0 instead of NaN/categoricaldata and transforming objects
label = LabelEncoder()

for c in df_train.columns:

    df_train[c]=df_train[c].fillna(0)

    if df_train[c].dtype == 'object':

        label.fit(list(df_train[c].values))

        df_train[c] = label.transform(list(df_train[c].values))

for c in df_test.columns:

    df_test[c]=df_test[c].fillna(0)

    if df_test[c].dtype == 'object':

        label.fit(list(df_test[c].values))

        df_test[c] = label.transform(list(df_test[c].values))
```

```python
#Split data into x,y of training and testing for prediction
x_train = df_train.drop(['parcelid', 'logerror', 'transactiondate', 'propertyzoningdesc', 'propertycountylandusecode', ], axis=1)
x_test = df_test.drop(['parcelid', 'propertyzoningdesc', 'propertycountylandusecode', '201610', '201611', '201612', '201710', '201711', '201712'], axis = 1)
x_train = x_train.values
y_train = df_train['logerror'].values
```

```python
#create liner model
linear = linear_model.LinearRegression()
#fit the model
linear.fit(x_train, y_train)
#score the model
linear.score(x_train, y_train)
```

```python
entire_predicted= linear.predict(x_test)
```

```python
#write entire prediction to csv
sample_file = pd.read_csv('sample_submission.csv')
```

```python
for c in sample_file.columns[sample_file.columns != 'ParcelId']:

    sample_file[c] = entire_predicted
```

```python
print('Writing to CSV')

sample_file.to_csv('predicted_results1.csv', index=False, float_format='%.4f')

print("Done")
```

```python
#Random Forest Regression
#Split data into x,y of training and testing for prediction
x_train = df_train[['calculatedfinishedsquarefeet', 'fips', 'taxamount', 'structuretaxvaluedollarcnt']]
x_test = df_test[['calculatedfinishedsquarefeet', 'fips', 'taxamount', 'structuretaxvaluedollarcnt']]
x_train = x_train.values
y_train = df_train['logerror'].values
```
